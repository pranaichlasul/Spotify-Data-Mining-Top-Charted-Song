## Data cleaning

### Subtask:
Data Cleaning: Handle inconsistencies and missing values, and remove outliers.


**Reasoning**:
The previous data exploration revealed several columns with non-numeric values that need to be cleaned.  The 'Streams' column and other numeric columns need to be converted to their correct data types.  I will also address missing values and outliers.



```python
# Convert 'Streams' to numeric, coercing errors to NaN
df['Streams'] = pd.to_numeric(df['Streams'], errors='coerce')

#Further cleaning for other numeric columns
for col in ['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Liveness', 'Tempo', 'Valence']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Handle missing values (imputation with the mean for numerical columns)
for col in ['Streams', 'Danceability', 'Energy', 'Loudness', 'Speechiness', 'Liveness', 'Tempo', 'Valence']:
    if df[col].isnull().any():
        df[col].fillna(df[col].mean(), inplace=True)

# Outlier detection and treatment (Winsorization for 'Streams')
for col in ['Streams', 'Danceability', 'Energy', 'Loudness', 'Speechiness', 'Liveness', 'Tempo', 'Valence']:
  upper_limit = df[col].quantile(0.95)
  lower_limit = df[col].quantile(0.05)
  df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

display(df.head())
display(df.info())
```

```python
# prompt: Lakukan Unsupervised Learning clurstering dengan K-means. clustering, pemilihan banyak cluster, menjelaskan hasil cluster (pemberian nama/identitas dari setiap cluster yang dihasilkan), dan penggunaan evaluasi

import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Select features for clustering (numerical features)
features = ['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Liveness', 'Tempo', 'Valence']
X = df[features]

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters using the Elbow Method
inertia = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()

# Based on the Elbow Method, let's choose a number of clusters (e.g., 3 or 4)
# Let's try with 3 clusters for demonstration
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Evaluate the clustering using Silhouette Score
silhouette_avg = silhouette_score(X_scaled, df['Cluster'])
print(f"\nSilhouette Score for {n_clusters} clusters: {silhouette_avg}")

# Analyze the characteristics of each cluster
print("\nCluster Analysis:")
cluster_means = df.groupby('Cluster')[features].mean()
display(cluster_means)

# Visualize the clusters (example using two features)
plt.figure(figsize=(10, 7))
plt.scatter(df['Danceability'], df['Energy'], c=df['Cluster'], cmap='viridis', s=50, alpha=0.6)
plt.title('K-Means Clustering (Danceability vs Energy)')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.colorbar(label='Cluster')
plt.show()

# Based on the 'cluster_means' output, you can interpret and name the clusters.
# For example:
# Cluster 0 might represent songs with low streams and moderate characteristics.
# Cluster 1 might represent songs with high streams and specific characteristics.
# Cluster 2 might represent songs with medium streams and other characteristics.
# You would analyze the average values of features within each cluster to give them meaningful names.
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your cleaned DataFrame from the previous steps

# Pastikan kolom 'Genre' ada di DataFrame Anda dan berisi data genre.
# Jika nama kolomnya berbeda, silakan ganti 'Genre' dengan nama kolom yang sesuai.
if 'Genre' not in df.columns:
    print("Error: Column 'Genre' not found in the DataFrame.")
    # Handle the error, maybe by stopping the execution or loading a different dataset
    # For demonstration, I will assume 'Genre' column exists.
    # You might need to add a step here to load/create the 'Genre' column if it's missing.
    # Example: df['Genre'] = ... based on your data structure
    pass # Placeholder, replace with actual error handling

# 1. Definisikan Variabel Independen (Fitur) dan Variabel Target
features = ['Energy', 'Acousticness', 'Tempo'] # Fitur yang akan digunakan
X = df[features]
y = df['Genre'] # Variabel target (Genre)

# 2. Encoding Variabel Target (Genre)
# Naive Bayes di scikit-learn memerlukan target berupa angka, jadi kita encode genre menjadi angka
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Mapping dari angka ke nama genre asli
genre_mapping = dict(zip(range(len(label_encoder.classes_)), label_encoder.classes_))
print("Mapping Genre:")
print(genre_mapping)

# Check class distribution
unique_counts = np.unique(y_encoded, return_counts=True)
print("\nClass distribution before filtering:")
for class_label, count in zip(unique_counts[0], unique_counts[1]):
    print(f"Class {class_label} ({label_encoder.inverse_transform([class_label])[0]}): {count} samples")

# Remove classes with only one sample
min_samples_per_class = 2
mask = np.isin(y_encoded, np.where(unique_counts[1] >= min_samples_per_class)[0])
X_filtered = X[mask]
y_encoded_filtered = y_encoded[mask]

print("\nClass distribution after filtering:")
unique_counts_filtered = np.unique(y_encoded_filtered, return_counts=True)
for class_label, count in zip(unique_counts_filtered[0], unique_counts_filtered[1]):
    print(f"Class {class_label} ({label_encoder.inverse_transform([class_label])[0]}): {count} samples")

# 3. Bagi Data Menjadi Data Training dan Testing
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y_encoded_filtered, 
                                                   test_size=0.25, random_state=42, 
                                                   stratify=y_encoded_filtered)

# 4. Bangun dan Latih Model Naive Bayes (Gaussian Naive Bayes untuk data kontinu)
model = GaussianNB()
model.fit(X_train, y_train)

# 5. Prediksi pada Data Testing
y_pred = model.predict(X_test)
y_pred_prob = model.predict_proba(X_test) # Probabilitas prediksi

# 6. Evaluasi Model
print("\nEvaluasi Model Naive Bayes:")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Visualisasikan Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted Genre')
plt.ylabel('Actual Genre')
plt.title('Confusion Matrix')
plt.show()

# Classification Report (Akurasi, Presisi, Recall, F1-Score per kelas)
print("\nClassification Report:")
# Get unique classes in the filtered data
unique_classes = np.unique(y_encoded_filtered)
filtered_class_names = label_encoder.classes_[unique_classes]
print(classification_report(y_test, y_pred, target_names=filtered_class_names))

# Akurasi Keseluruhan
accuracy = accuracy_score(y_test, y_pred)
print(f"\nOverall Accuracy: {accuracy:.4f}")

# Catatan tentang Goodness of Fit untuk Naive Bayes:
# Naive Bayes tidak memiliki konsep "koefisien" seperti regresi logistik.
# Goodness of Fit dievaluasi terutama melalui metrik kinerja pada data testing,
# seperti akurasi, presisi, recall, F1-score, dan confusion matrix.
# ROC-AUC dapat dihitung, tetapi lebih umum untuk masalah biner. Untuk multi-kelas,
# Anda bisa menghitung macro-average atau micro-average AUC, tetapi classification report
# lebih informatif untuk kinerja per kelas.

# 7. Interpretasi Hasil
print("\nInterpretasi Hasil:")
print("- Confusion Matrix: Menunjukkan jumlah prediksi yang benar dan salah untuk setiap pasangan genre (aktual vs prediksi).")
print("  - Diagonal matriks menunjukkan prediksi yang benar.")
print("  - Nilai di luar diagonal menunjukkan kesalahan klasifikasi.")
print("- Classification Report:")
print("  - Precision: Kemampuan model untuk tidak mengklasifikasikan sampel negatif sebagai positif untuk setiap genre.")
print("  - Recall: Kemampuan model untuk menemukan semua sampel positif untuk setiap genre.")
print("  - F1-Score: Rata-rata harmonik dari Precision dan Recall, memberikan keseimbangan.")
print("  - Support: Jumlah sampel aktual untuk setiap genre di set testing.")
print("- Overall Accuracy: Proporsi total prediksi yang benar.")
print("- Probabilitas Prediksi (`y_pred_prob`): Menunjukkan probabilitas sampel termasuk dalam setiap kelas genre. Genre dengan probabilitas tertinggi adalah prediksi model.")

# Contoh prediksi untuk data baru (dengan nilai Energy, Acousticness, Tempo)
# Ganti nilai-nilai ini dengan data lagu yang ingin Anda prediksi genrenya
new_song_features = [[0.7, 0.2, 120]] # Contoh: Energy=0.7, Acousticness=0.2, Tempo=120

# Pastikan fitur input dalam format yang sama dengan data training (misalnya, numpy array atau list of lists)
predicted_genre_encoded = model.predict(new_song_features)
predicted_genre = label_encoder.inverse_transform(predicted_genre_encoded)

predicted_genre_prob = model.predict_proba(new_song_features)

print(f"\nContoh Prediksi untuk Lagu Baru {new_song_features}:")
print(f"Diprediksi Genre: {predicted_genre[0]}")
print(f"Probabilitas Prediksi per Genre: {predicted_genre_prob}")

# Menampilkan probabilitas dengan nama genre yang sesuai
prob_dict = dict(zip(label_encoder.classes_, predicted_genre_prob[0]))
print("Probabilitas:")
for genre, prob in prob_dict.items():
    print(f"- {genre}: {prob:.4f}")
```

### 1. Use More Features

We will now use additional features: `Danceability`, `Loudness`, `Speechiness`, and `Valence` along with the previous features for genre prediction.

```python
# Use more features for genre prediction
selected_features = [
    'Energy', 'Acousticness', 'Tempo',
    'Danceability', 'Loudness', 'Speechiness', 'Valence'
]
X_full = df[selected_features]
y_full = df['Genre']

print("Selected features for modeling:", selected_features)
display(X_full.head())
```

### 2. Feature Engineering

Let's create new features such as ratios, interactions, or binning. For example, we can create:
- `Energy_to_Loudness`: Ratio of Energy to Loudness (avoid division by zero)
- `Is_Fast_Tempo`: 1 if Tempo > median, else 0
- `Danceability_Squared`: Square of Danceability

```python
# Feature engineering
X_fe = X_full.copy()
X_fe['Energy_to_Loudness'] = X_fe['Energy'] / (X_fe['Loudness'].replace(0, 1))
X_fe['Is_Fast_Tempo'] = (X_fe['Tempo'] > X_fe['Tempo'].median()).astype(int)
X_fe['Danceability_Squared'] = X_fe['Danceability'] ** 2

print("Feature engineered columns:", X_fe.columns.tolist())
display(X_fe.head())
```

### 3. Try More Powerful Models

We will try Random Forest, SVM, and XGBoost classifiers and compare their accuracy using cross-validation.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder

# Encode target
le = LabelEncoder()
y_encoded = le.fit_transform(y_full)

# Random Forest
rf = RandomForestClassifier(random_state=42)
rf_score = cross_val_score(rf, X_fe, y_encoded, cv=5, scoring='accuracy').mean()

# SVM (linear kernel)
svm = SVC(kernel='linear', random_state=42)
svm_score = cross_val_score(svm, X_fe, y_encoded, cv=5, scoring='accuracy').mean()

# XGBoost
xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
xgb_score = cross_val_score(xgb, X_fe, y_encoded, cv=5, scoring='accuracy').mean()

print(f"Random Forest CV Accuracy: {rf_score:.2f}")
print(f"SVM (linear) CV Accuracy: {svm_score:.2f}")
print(f"XGBoost CV Accuracy: {xgb_score:.2f}")
```
